```python
import seaborn as sns
import numpy as np
import pandas as np
```

```python
panguins['island'].value_counts()
```

```python
sns.scatterplot(data=panguins, x="flipper_length_mm",y="body_mass_g",hue="species")
```

```python
 sns.scatterplot(data=panguins, x="flipper_length_mm",y="body_mass_g",hue="species", palette="Set1")
```
