```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y=[10,25,5,15,12,25,13,6,9]
#plt.plot(x,y, color='',linestyle='',linewidth=value,marker='marker-symbol',lable='name')
#only x,y are must given and less are optional
plt.plot(x,y,color="green",linestyle='--',marker="o",label="1st work")
plt.title("Basic Understanding")
plt.xlabel("Value of X ")
plt.ylabel("Value of Y")
plt.legend()
plt.grid(color="gray",linestyle="--",linewidth=1)#parameter is optional

plt.show()
```

```python
#plt.bar
product=['A','B','C','D']
sales=[1200,1500,1800,1300]
plt.bar(product,sales, color="orange",label="sales 2025")
#plt.barh->horizontal
plt.xlabel("Product")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
Region=["North","South","East","West"]
Revinue=[3000,5000,2000,8000]
plt.pie(Revinue,labels=Region,autopct='%1.1f%%',colors=['gold','skyblue','black','pink'])
plt.title("Revenue")
plt.show()
```

```python
score=[11,34,33,45,27,46,34,63,56,63,55]
plt.hist(score,bins=5,color='pink')
plt.show()
```

```python
hour_studied=[1,2,3,4,5,6,7]
exam_score=[50,55,70,65,45,66,95]
plt.scatter(hour_studied,exam_score,color="red",marker="o")
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,10,40]
plt.subplot(1,2,1) #(nrow,ncol,index)
plt.plot(x,y)
plt.title("linechart")
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("barchat")
plt.show()
```
